# Question 1 [35 Points] Local Linear Regression

We have implemented the Nadaraya-Watson kernel estimator in HW 6. In this question, we will investigate a local linear regression:
$$
\widehat{f}\left(x\right)=\widehat{\beta}_{0}\left(x\right)+ \widehat{\beta}_{1}\left(x\right) x,
$$
where $x$ is a testing point. Local coefficients $\widehat{\beta}_{r}\left(x \right)$ for $r=0, 1$ are obtained by minimizing the object function
$$
\underset{\beta_{0}(x), \, \beta_{1}(x)}{\operatorname{minimize}} \quad \sum_{i=1}^{n} K_{\lambda} \left(x, x_{i}\right) \Big[y_{i}-\beta_{0}(x) - \beta_1(x) x_{i} \Big]^{2}.
$$

In this question, we will use the Gaussian kernel $K(u) = \frac{1}{\sqrt{2 \pi}} e^{- \frac{u^2}{2}}$.

a) [20 pts] Write a function `myLocLinear(trainX, trainY, testX, lambda)`, where `lambda` is the bandwidth and `testX` is all testing samples. This function returns predictions on `testX`. The solution of $\beta_{0}(x)$ and $\beta_{1}(x)$ can be obtained by fitting a weighted linear regression. The formula is provided on Page 25 of our [lecture note](https://teazrq.github.io/stat542/notes/Kernel.pdf). 

b) [15 pts] Fit a local linear regression with our given training data. The testing data are generated using the code given below. Try a set of bandwidth  $\lambda = 0.05, 0.1, \ldots, 0.55, 0.6$ when calculating the kernel function.

  - Report the best testing MSE with the corresponding $\lambda$.

```python
### Import Packages ###
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import math
from tqdm import tqdm
from sklearn.metrics import mean_squared_error

# Read in Data
data = pd.read_csv("hw7_Q1_train.csv")

X_train = data["x"].values
y_train = data["y"].values

X_test = 2 * math.pi * np.arange(0, 1, 0.01)
y_test = [math.sin(x) for x in X_test]
```

```python
class LocalWeightedRegression:
    def __init__(self, X, y, X_test, k):
        self.X = X.reshape(-1, 1)
        self.y = y.reshape(-1, 1)
        self.X_test = X_test.reshape(-1, 1)
        self.k = k
        self.n, self.p = self.X.shape
        self.n_test, self.p_test = self.X_test.shape
        
    def gaussian_kernel(self, target_x):
        weight_diag = np.eye((self.n))
        gauss_kernel = lambda x: (1/(2*math.pi)**0.5)*math.exp((-x**2)/2)
        for i in range(self.n):
            distance = abs(target_x - self.X[i])
            weight_diag[i, i] = gauss_kernel(distance/self.k)/self.k
    
        return weight_diag

    def local_weights(self, target_x):
        w = self.gaussian_kernel(target_x)
        weight = np.linalg.inv(self.X.T @ (w @ self.X)) @ (self.X.T @ w @ self.y)
        return weight
        
    def predict(self, train=True):
        if train:
            y_prediction = np.zeros(self.n)
            for i in range(self.n):
                y_prediction[i] = self.X[i] * self.local_weights(self.X[i])
            return y_prediction
        else:
            y_prediction = np.zeros(self.n_test)
            for i in range(self.n_test):
                y_prediction[i] = self.X_test[i] * self.local_weights(self.X_test[i])
            return y_prediction
```

```python
lambdas = np.arange(0.05, 0.65, 0.05)
mse_list = []
for l in lambdas:
    y_pred = LocalWeightedRegression(X_train, y_train, X_test, k=l).predict(train=False)
    mse = mean_squared_error(y_pred, y_test)
    mse_list.append(mse)
    
plt.plot(lambdas, mse_list)
plt.xlabel("Lambda Values")
plt.ylabel("Mean Squared Error")
plt.title("Lambda vs MSE for Gaussian Weighted Linear Regression")
plt.plot()
```

```python
y_pred_05 = LocalWeightedRegression(X_train, y_train,X_test, k=0.05).predict(train=False)
y_pred_25 = LocalWeightedRegression(X_train, y_train,X_test, k=0.25).predict(train=False)
y_pred_50 = LocalWeightedRegression(X_train, y_train,X_test, k=0.5).predict(train=False)

fig, axs = plt.subplots(3, figsize=(15,20))

axs[0].plot(X_test, y_pred_05, color="r", lw=5, label="Predicted")
axs[0].plot(X_test, y_test, color="y", lw=5, label="Truth")
axs[0].scatter(X_train,y_train)
axs[0].legend()
axs[0].set_title("Lambda: 0.05")

axs[1].plot(X_test, y_pred_25, color="r", lw=5, label="Predicted")
axs[1].plot(X_test, y_test, color="y", lw=5, label="Truth")
axs[1].scatter(X_train,y_train)
axs[1].legend()
axs[1].set_title("Lambda: 0.25")

axs[2].plot(X_test, y_pred_25, color="r", lw=5, label="Predicted")
axs[2].plot(X_test, y_test, color="y", lw=5, label="Truth")
axs[2].scatter(X_train,y_train)
axs[2].legend()
axs[2].set_title("Lambda: 0.5")
```

  a. [10 pts] Estimate the mean, covariance matrix of each class and pooled covariance matrix. Basic built-in R functions such as `cov` are allowed. Do NOT print your results. 
  b. [15 pts] Write your own linear discriminate analysis (LDA) code following our lecture note. To perform this, you should calculate $\mu_k$, $\pi_k$, and $\Sigma$ from the data. You may consider saving $\mu_k$'s and $\pi_k$'s as a list (with 10 elements in each list). 


```python
# Load Data
digits_train = pd.read_csv("digits_train.csv")
digits_test = pd.read_csv("digits_test.csv")

digits_train.head()

X_train = digits_train.drop(columns="V1")
X_test = digits_test.drop(columns="V1")

y_train = digits_train["V1"]
y_test = digits_test["V1"]
```

```python
count = np.array(y_train.value_counts())
class_mean = np.array([group.values[:, 1:].mean(axis=0) for grouper, group in digits_train.groupby("V1")])
class_cov = np.array([group.iloc[:, 1:].cov() for grouper, group in digits_train.groupby("V1")])
pooled_cov = np.sum([count[i]/(y_train.shape[0] - y_train.nunique())*class_cov[i] 
                     for i in range(class_cov.shape[0])], axis=0)


def LDA():
    predictions = []
    correct = 0
    pooled_inverse = np.linalg.inv(pooled_cov)
    conf_mat = np.zeros(shape=(y_test.nunique(),y_test.nunique()))
    for idx, row in tqdm(digits_test.iterrows(), total=len(digits_test)):
        label = int(row.V1)
        delta = np.zeros(y_test.nunique())
        for i in range(len(delta)):
            delta[i] = class_mean[i].T@pooled_inverse@X_test.values[idx]
            delta[i] = delta[i] - 0.5*(class_mean[i].T@pooled_inverse@class_mean[i]) + np.log(count[label]/y_test.shape[0])
        
        if delta.argmax() == y_test[idx]:
            correct += 1
        predictions.append(delta.argmax())
        conf_mat[delta.argmax(), y_test[idx]] += 1
    accuracy = correct / len(y_test)
    return accuracy, predictions, conf_mat
        
accuracy, predictions, conf_mat = LDA()
print("Accuracy: {}".format(accuracy))

weights = np.linalg.inv(pooled_cov)@(class_mean[0].reshape(-1,1))
print("First 5 Weights for Number 0: {}".format(weights[:5].reshape(-1)))

bias = np.log(count[0]/y_train.shape[0]) - 0.5 * (class_mean[0].T@np.linalg.inv(pooled_cov)@class_mean[0])
print("Bias for Number 0: {}".format(bias))
```

```python
error_rate = (conf_mat.T / conf_mat.sum(axis=0)).T
for i in range(len(error_rate)):
    print("Error Rate for Number {}: {}".format(i, round(1 - error_rate[i,i], 3)))

print("Overall Error Rate: {}".format(1 - accuracy))
```
